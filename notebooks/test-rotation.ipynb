```python
from preprocess import correct_orientation
import os
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
cur_path = os.path.dirname(os.getcwd())
data_path = os.path.join(cur_path, "sports_scr/data/natation/")
path_tuples = [(f"00000{i}.png", f"00000{i}.txt") for i in range(10)]
path_tuples.extend([(f"0000{i}.png", f"0000{i}.txt") for i in range(10, 15)])
```

```python
for i in range(15):
    im_path = data_path + path_tuples[i][0]
    img = np.array(Image.open(im_path))
    rotated = correct_orientation(img)
    plt.figure()
    _, ax = plt.subplots(1, 2)
    ax[0].imshow(img)
    ax[1].imshow(rotated)
    plt.show()
```
